```python
Y_val=data.iloc[:,0].values
```

```python
z=np.zeros((1,1))
for i in Y_val:
    if i=='ham':
        z=np.vstack((z,1))
    else:
        z=np.vstack((z,0))
z=z[1:,0]
```

```python
Y_val=z
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize()
```

```python
token=[]
for i in mssg:
    z=word_tokenize(i,language='english')
    print(z)
    token.append(z)
```

```python
token
```

```python
from nltk.corpus import stopwords
s_words=list(stopwords.words('English'))
s_words
```

```python
import re
```

```python
stage1=[]
for i in token:
    lst=[]
    for j in i:
        if re.search('^[\.\,\\!&0-9\'å\>\<\:\?\-\;\#\@\*\)\(]',j):
            continue
        elif j in s_words:
            continue
        else:
            if not j in lst:
                lst.append(j)
    stage1.append(lst)
        
        
```

```python
stage1
```

```python
## Stemming & Lemmatization
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
stage2=[]
for i in stage1:
    lst=[]
    for j in i:
        word=wnl.lemmatize(j)
        if not word in lst:
            lst.append(word)
    stage2.append(lst)
```

```python
stage2
```

```python
## Count-Vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
stage2_1=[]
for i in stage2:
    sent=' '.join(i)
    stage2_1.append(sent)
stage2_1
        
```

```python
arr=cv.fit_transform(stage2_1).toarray()
```

```python
arr
```

```python
index=[]
for i in range(len(Y_val)):
    if Y_val[i]==0:
        index.append(i)
```

```python
index
```

```python
arr_xcut=arr[index,:]
```

```python
arr_xcut
```

```python
header=cv.vocabulary_.items()
header=sorted(header,key= lambda f:f[1],reverse=False)
```

```python
header
```

```python
check=arr_xcut.sum(axis=0)
c=pd.DataFrame(check)
c.to_csv('checking.csv')
```

```python
checkl=list(check)
h_v=[]
for i in range(len(checkl)):
    if checkl[i]>10:
        h_v.append(i)
h_v
```

```python
arr_xcut=arr_xcut[:,h_v]
```

```python
arr_xcut.sum(axis=0)
```

```python
f_list=[]
for i in h_v:
    f_list.append(header[i][0])
f_list
```

```python
def check_up(doc):
#doc="SIX chances to win CASH! From 100 to 20,000 pounds txt> CSH11 and send to 87575. Cost 150p/day, 6days, 16+ TsandCs apply Reply HL 4 info"
    tt=word_tokenize(doc, language='english')
    lst2=[]
    for i in tt:
        if re.search('^[\.\,\\!&0-9\'å\>\<\:\?\-\;\#\@\*\)\(]',i.lower()):
            #print("Yeah")
            continue
        if not i.lower() in lst2:
            lst2.append(i.lower())
    print(lst2)
    lst4=[]
    for i in lst2:
        if not i in s_words:
            lst4.append(i)
    print(lst4)
    wnl = WordNetLemmatizer()
    lst8=[]
    for i in lst4:
        zz=wnl.lemmatize(i)
        if not zz in lst8:
            lst8.append(zz)
    print(lst8)
    count=0
    for i in lst8:
        if i in f_list:
            count=count+1
            print(i)
    f_ans=(count/len(lst8))
    if f_ans>=0.5:
        return(1)
    else:
        return(0)
```

```python
tally=[]
for i in mssg:
    val=check_up(i)
    tally.append(val)
```

```python
tally
```

```python
Y_new=[]
for i in Y_val:
    if i==0:
        Y_new.append(1)
    else:
        Y_new.append(0)
```

```python
sount=0
for i in range(len(tally)):
    if Y_new[i]==tally[i]:
        sount=sount+1
print(sount/len(tally))
```
