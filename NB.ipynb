```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
corpus1=pd.read_csv('spam.csv')
corp=corpus1.values[:,1].reshape((-1,))
t=int(0.6*corp.shape[0])
corpus=corp[:t]
Y_=corpus1.values[:,0].reshape((-1,))
Y_val=Y_[:t]
dd=corp[t:]
dd
```

```python
tok=[]
for i in corpus:
    wt=word_tokenize(i,language='english')
    tok.append(wt)
```

```python
tok
```

```python
from nltk.corpus import stopwords
```

```python
sw=stopwords.words('English')
```

```python
import re
```

```python
clean=[]
for i in tok:
    lst=[]
    for j in i:
        j=j.lower()
        if re.search('^[\.\,\\!&0-9\'å\>\<\:\?\-\;\#\@\*\)\(]',j):
            continue
        if not j in sw:
            lst.append(j)
    clean.append(lst)
clean
```

```python
## cleaning done
```

```python
from nltk.stem import WordNetLemmatizer 
```

```python
from nltk.stem import LancasterStemmer
```

```python
wnl=WordNetLemmatizer()
st=LancasterStemmer()
```

```python
lem=[]
for i in clean:
    lst=[]
    for j in i:
        zz=wnl.lemmatize(j)
        lst.append(zz)
    lem.append(lst)
lem
```

```python
## lemmatized and ready for vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
data=[]
for i in lem:
    data.append(' '.join(i))
data
```

```python
arr=cv.fit_transform(data).toarray()
```

```python
## Bernoulli NB implemented from scratch
def prior(fmat,label):
    z=fmat[:,-1]
    a=z[z[:]==label]
    return len(a)/len(z)

def likelihood(arr,f_index,label):
    z=arr[:,-1]
    #print(label,f_index,z,'likelihood')
    like=1.0
    a=z[z[:]==label]
    #print(a,'a')
    np1=arr[:,f_index][z[:]==label]
    #print(np1,'np1')
    like=like*(((np1.sum(axis=0)+1)/(len(a)+arr.shape[1])))
    #print(like,'@')
    return like
    
def posterior(arr,Y_val,lstream):
    fmat=np.hstack((arr,Y_val))
    lab=np.unique(Y_val)
    post=[]
    for i in list(lab):
        e=1.0
        for j in range(len(lstream)):
            #print(lstream)
            l=likelihood(fmat,j,i)
            p=prior(fmat,i)
            #print(l,p,"2 vars")
            #print(lstream[j],'ls')
            e=e*((l**lstream[j])*((1-l)**(1-lstream[j])))
            #print(e)
        e=e*p
        #print(e,"error")
        ntup=(e,i)
        post.append(ntup)
        #print(post,"post")
    #print('post_final')
    lol=sorted(post,reverse=True)[0]
    #print(lol,'ans')
    print(lol[1])
    return lol[1]
```

```python
def check_up2(doc):
    tt=word_tokenize(doc, language='english')
    lst2=[]
    for i in tt:
        if re.search('^[\.\,\\!&0-9\'å\>\<\:\?\-\;\#\@\*\)\(]',i.lower()):
            #print("Yeah")
            continue
        if not i.lower() in lst2:
            lst2.append(i.lower())
    #print(lst2)
    lst4=[]
    for i in lst2:
        if not i in sw:
            lst4.append(i)
    #print(lst4)
    wnl = WordNetLemmatizer()
    lst8=[]
    for i in lst4:
        zz=wnl.lemmatize(i)
        if not zz in lst8:
            lst8.append(zz)
    #print(lst8)
    return (lst8)

dd1=[]
for i in dd:
    test_corpus=i
    dd1.append(check_up2(test_corpus))
dd1
```

```python
iot=[]
for dd in dd1:
    iot.append(' '.join(dd))
iot
```

```python
f_list=[]
for i in range(len(iot)):
    kkg=cv.transform([iot[i]]).toarray()
    ans=posterior(arr,Y_val,kkg[0])
    f_list.append(ans)
f_list
```
